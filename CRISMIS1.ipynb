```python
import requests
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
# url="https://pdsimage2.wr.usgs.gov/archive/mess-e_v_h-mdis-2-edr-rawdata-v1.0/MSGRMDS_1001/DATA/2011_207/EW0220137668G.IMG"
url = "https://pdsimage2.wr.usgs.gov/archive/mess-e_v_h-mdis-2-edr-rawdata-v1.0/MSGRMDS_1001/DATA/2011_156/EN0215728629M.IMG"
r = requests.get(url, allow_redirects=True)
open('google.IMG', 'wb').write(r.content)

input_filename = "google.IMG"
```

```python
# shape = (1031, 512)
from PIL import Image
from pylab import *
dtype = np.dtype('>u2')
fid = open(input_filename, 'rb')
data = np.fromfile(fid, dtype)
print(np.shape(data))
da=np.shape(data)
p=int(da[0]/512)

shape = (p, 512)
image = data.reshape(shape)
# print(image)
plt.imshow(image, cmap = "gray")
axis("off")
# plt.savefig('df.png')
# plt.savefig("df.png",bbox_inches='tight')
# plt.savefig("df.png", transparent = True, bbox_inches = 'tight', pad_inches = 0)
# plt.show()
# plt.gca().set_axis_off()
# plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
#             hspace = 0, wspace = 0)
# plt.margins(0,0)
# plt.savefig("myfig.pdf")
# plt.axis("off")
# fig=plt.imshow(image,interpolation='nearest')
# fig.axes.get_xaxis().set_visible(False)
# fig.axes.get_yaxis().set_visible(False)
# plt.savefig('df.png',
#     bbox_inches='tight', pad_inches=0, format='png', dpi=1200)
plt.imsave("dff.png",image,cmap="gray")
```

```python
import cv2

image = cv2.imread('dff.png')

alpha = 3 # Contrast control (1.0-3.0)
beta = 0# Brightness control (0-100)

adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

cv2.imshow('original', image)
cv2.imshow('adjusted', adjusted)
cv2.waitKey()
```
